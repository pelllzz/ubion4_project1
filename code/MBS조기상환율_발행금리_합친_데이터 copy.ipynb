```python
# 전처리를 위한 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
# MBS조기상환율 데이터 불러오기
mbs1 = pd.read_csv("./datasets/MBS_200406-200412.csv")
mbs2 = pd.read_csv("./datasets/MBS_200501-201012.csv")
mbs3 = pd.read_csv("./datasets/MBS_201101-201512.csv")
mbs4 = pd.read_csv("./datasets/MBS_201601-202012.csv")
mbs5 = pd.read_csv("./datasets/MBS_202101-202208.csv")
```

```python
# 필요한 칼럼만 추출
mbs1 = mbs1[['발행일자', '발행회차', '기준월', '만기전상환율', '만기전상환금액']]
mbs2 = mbs2[['발행일자', '발행회차', '기준월', '만기전상환율', '만기전상환금액']]
mbs3 = mbs3[['발행일자', '발행회차', '기준월', '만기전상환율', '만기전상환금액']]
mbs4 = mbs4[['발행일자', '발행회차', '기준월', '만기전상환율', '만기전상환금액']]
mbs5 = mbs5[['발행일자', '발행회차', '기준월', '만기전상환율', '만기전상환금액']]
```

```python
#MBS 조기상환율 모든 연도 데이터 합치기
mbs = pd.concat([mbs1, mbs2, mbs3, mbs4, mbs5], ignore_index=True)
mbs.reset_index(drop=False, inplace=True)
mbs
```

```python
# MBS 조기상환율 발행일자, 발행회차 기준 평균내기
mbs = mbs.groupby(["index", "발행일자", "발행회차"])['만기전상환율'].apply(np.average)
mbs = pd.DataFrame(mbs)
mbs
```

```python
# 발행금리 불러오기
r = pd.read_csv('./datasets/MBS발행금리_20040101-20220810.csv')
r = r[['발행회차', '발행일자', '가중평균발행금리']]
```

```python
# MBS 조기상환율 & 발행금리 합치기
mbs_r = pd.merge(mbs, r, how='left', on=['발행일자', '발행회차'])
mbs_r
```

```python
# 팔행일자, 발행회차 기준 데이터 정리
mbs_r = mbs_r.set_index(['발행일자', '발행회차'])
mbs_r
```
